```python
random.seed(42)

def generate_data(n_samples):
    """
    Generate linear data: y = 3*x1 + 2*x2 + noise
    """
    X = []
    y = []
    for _ in range(n_samples):
        x1 = random.uniform(-10, 10)
        x2 = random.uniform(-10, 10)
        noise = random.uniform(-0.5, 0.5)
        X.append([x1, x2])
        y.append(3 * x1 + 2 * x2 + noise)
    return X, y

def benchmark():
    sizes = [10, 100, 1000, 5000, 10000, 20000, 50000, 100000, 500000, 1000000, 5000000, 10000000]
    cython_times = []
    sklearn_times = []

    for size in sizes:
        X, y = generate_data(size)

        # Cython Timing
        start = time.time()
        fit_linear_regression(X, y)
        cython_times.append((time.time() - start) * 1e3)  # ms

        # Sklearn Timing
        start = time.time()
        model = LinearRegression(fit_intercept=False)
        model.fit(X, y)
        sklearn_times.append((time.time() - start) * 1e3)  # ms

        print(f"Size: {size:5d} | Cython: {cython_times[-1]:7.3f} ms | Sklearn: {sklearn_times[-1]:7.3f} ms")

    return sizes, cython_times, sklearn_times

def plot(sizes, cython_times, sklearn_times):
    plt.figure(figsize=(10, 6))
    plt.plot(sizes, cython_times, 'o--', label='Cython Linear Regressor', color='blue')
    plt.plot(sizes, sklearn_times, 's--', label='Sklearn LinearRegression', color='green')
    plt.title("Performance Comparison")
    plt.xlabel("Number of Samples")
    plt.ylabel("Time (ms)")
    plt.grid(True, linestyle='-', alpha=0.8)
    plt.legend()
    plt.tight_layout()
    plt.show()

def log_plot(sizes, cython_times, sklearn_times):
    plt.figure(figsize=(10, 6))
    plt.plot(sizes, cython_times, 'o--', label='Cython Linear Regressor', color='blue')
    plt.plot(sizes, sklearn_times, 's--', label='Sklearn LinearRegression', color='green')
    plt.title("Performance Comparison")
    plt.xlabel("Number of Samples")
    plt.ylabel("Time (ms)")
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.yscale('log')
    plt.legend()
    plt.tight_layout()
    plt.show()

# Run
sizes, cython_times, sklearn_times = benchmark()
plot(sizes, cython_times, sklearn_times)
log_plot(sizes, cython_times, sklearn_times)
```
